```python
"""
Question-1. Dr. Alam aims to develop an automated system for diabetes prediction and seeks assistance
from a machine learning specialist to build the AI model. He has provided a dataset available at:
https://drive.google.com/file/d/1BAtm0a41zOuWogkcmJnJcJyWGdp62_h1
Using this dataset, perform the following tasks. Clearly show your code, outputs, and brief
explanations.
i. Demonstrate data exploration:
a. Display the first 10 rows of the dataset.
b. State the number of rows and columns.
c. Identify the target variable.
ii. Apply preprocessing steps appropriately:
a. Handle any missing values.
b. Encode categorical features if present.
c. Detect outliers or skewness and explain their treatment.
d. Split the dataset into training and test sets (e.g., 75/25 split) using a fixed random
seed.
e. Apply feature scaling/standardization if appropriate.
iii. Implement a Logistic Regression model to predict diabetes and report the results.
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

df = pd.read_csv('/content/diabetes.csv')
#Display the first 10 rows of the dataset.
print(df.head(10))
#State the number of rows and columns.
print(df.shape)
#Identify the target variable.
print(df.columns)
print(df['Outcome'].values)

#Handle any missing values.
print(df.isnull().sum())

#Encode categorical features if present.
print(df.dtypes)
#Detect outliers or skewness and explain their treatment.
print(df.describe())

#Split the dataset into training and test sets (e.g., 75/25 split) using a fixed random seed.
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state = 42)

#Apply feature scaling/standardization if appropriate.
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

#Implement a Logistic Regression model to predict diabetes and report the results.
model = LogisticRegression()
model.fit(x_train, y_train)
y_prediction = model.predict(x_test)
print(y_prediction)



```

```python
"""
Question-2. You are provided with a dataset on red wine quality, which includes chemical properties (like
acidity, pH, sugar, alcohol, etc.) and a quality score ranging from 0 (low) to 10 (high). The
dataset can be accessed here:
https://drive.google.com/file/d/1RNoCENA1qQCd3ML0EG1GxOacbnEGVjXJ
i. Preprocessing the Dataset:
a. Explain how missing values are identified and handled (if present).
b. Interpret the need for standardization or normalization of numeric features for
consistency.
c. Differentiate the quality score into three groups (low, medium, high) and Define your
thresholds for classification labels clearly.
ii. Classification and Evaluation:
Using a Naive Bayes classifier, train a model to predict wine quality categories.
Summarize the classifier’s performance using suitable metrics (e.g., accuracy, confusion
matrix, precision, recall).

"""

datasets = pd.read_csv('/content/winequality-red.csv')
print(datasets.head(3))
#Explain how missing values are identified and handled (if present).
print(datasets.isnull().sum())

#Interpret the need for standardization or normalization of numeric features for consistency.
print(datasets.describe())
#Differentiate the quality score into three groups (low, medium, high) and Define your thresholds for classification labels clearly.
print(datasets.columns)
print(datasets['quality'].values)
print("Cluster Group are: \n",datasets['quality'].unique())
print("\n")
print("Cluster Counts: ",datasets['quality'].value_counts())

"""
ii. Classification and Evaluation:
Using a Naive Bayes classifier, train a model to predict wine quality categories.
Summarize the classifier’s performance using suitable metrics (e.g., accuracy, confusion
matrix, precision, recall).
"""
x = datasets.iloc[:,:-1].values
y = datasets.iloc[:,-1].values
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state = 42)

#Apply feature scaling/standardization if appropriate.
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
y_prediction = model.predict(x_test)
print(y_prediction)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_prediction)
print("\n")
print("Confusion Metrics: \n",cm)
print("Accuracy: ",accuracy_score(y_test, y_prediction))

print("\n\n")
from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB()
model.fit(x_train, y_train)
y_prediction = model.predict(x_test)
print(y_prediction)

from sklearn.metrics import confusion_matrix, accuracy_score
cmm = confusion_matrix(y_test, y_prediction)
print("\n")
print("Confusion Metrics: \n",cmm)
print("Accuracy: ",accuracy_score(y_test, y_prediction))





```
