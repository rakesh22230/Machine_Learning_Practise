```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/rakesh22230/Machine_Learning_Practise/refs/heads/main/house_price_regression_dataset.csv")
# Display: first 5 rows; column names and data types;
df.head()
# missing values count.



df.head()

# Replace the first rows Square_Footage with the maximum value from the column.
df.isnull().sum()
df.info()

# Create a scatter plot for Square_Footage vs House_Price.
plt.scatter(df['Square_Footage'],df['House_Price'])
plt.title('Square_Footage VS Houese Price')
plt.xlabel('Square_Footage')
plt.ylabel('House_Price')
plt.show()

#Use Linear Regression to predict the house price.
x = df.iloc[:,:-1].values
y = df.iloc[:,:-1].values

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)

y_pred = regressor.predict(x_test)

# Evaluate using accuracy.
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)

df = pd.DataFrame({'Actual':y_test.flatten(),'Predicted':y_pred.flatten()})
df.head()





```
